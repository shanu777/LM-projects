```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.drop('subject#',inplace=True,axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
cols=df.columns
for m in cols:
    df[m]=ss.fit_transform(df[m].values.reshape(-1,1))
```

```python
df.corr()
```

```python
from sklearn.decomposition import PCA
```

```python
y=df.iloc[:,3:5].values.reshape(-1,2)
target=['motor_UPDRS','total_UPDRS']
```

```python
x=df.drop(target,axis=1)
```

```python
pca=PCA(n_components=16)
```

```python
x_pca=pca.fit_transform(x)
```

```python
x.shape
```

```python
x_pca.shape
```

```python
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_pca,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtrain,ytrain)
```

```python
lr_pred=lr.predict(xtest)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
```

```python
rfr.fit(xtrain,ytrain)
```

```python
rfr.score(xtrain,ytrain)
```

```python
rfr_pred=rfr.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(rfr_pred,ytest)
```

```python
mean_squared_error(lr_pred,ytest)
```

```python
r2_score(lr_pred,ytest)
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(output_dim=30,init='uniform',input_dim=16,activation='relu'))
classifier.add(Dense(output_dim=20,init='uniform',activation='relu'))
classifier.add(Dense(output_dim=2,init='uniform',activation='linear'))
classifier.compile(optimizer='adam',loss='mse',metrics=['mse'])
classifier.fit(xtrain,ytrain,batch_size=5,nb_epoch=200)
```

```python
kera_pred=classifier.predict(xtest)
```

```python
r2_score(kera_pred,ytest)
```
