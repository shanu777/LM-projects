```python
from imblearn.combine import SMOTETomek as smk
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
le=LabelEncoder()
on=OneHotEncoder()
```

```python
cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety','class']
```

```python
for m in cols:
    df[m]=le.fit_transform(df[m].values.reshape(-1,1))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x=df.drop('class',axis=1)
y=df.iloc[:,-1].values.reshape(-1,1)
```

```python
smk=smk()
```

```python
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,stratify=y)
```

```python
xtrain,ytrain=smk.fit_sample(xtrain,ytrain)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtrain,ytrain)
```

```python
accuracy_score(lr.predict(xtest),ytest)
```

```python
print(classification_report(lr.predict(xtest),ytest))
```

```python
rfr=RandomForestClassifier()
```

```python
rfr.fit(xtrain,ytrain)
```

```python
rfr.score(xtrain,ytrain)
```

```python
rfr.score(xtest,ytest)
```

```python
accuracy_score(rfr.predict(xtest),ytest)
```

```python
print(classification_report(rfr.predict(xtest),ytest))
```

```python
cv=cross_val_score(rfr,x,y,cv=10)
```
